 # *Define data visualization*

### Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data

# *Importance of Data Visualization*


### -Makes complex data accessible and understandable.
### -Helps identify trends and patterns quickly.
### -Supports decision-making processes.
### -Enhances communication and storytelling.

# *Types of Data Visualization*

#### Bar Charts
#### Line Charts
#### Pie Charts
#### Scatter Plots
#### Heatmaps
#### Treemaps
#### Infographics
#### Maps

# *Choosing the Right Visualization*

#### Type of data (categorical, numerical, time-series).
#### Purpose of the visualization (comparison, distribution, correlation).
#### Audience and context.

# Examples using paython libs

## 1-Bar Charts (Matplotlib)

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Plot
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

## 2-Line Charts (Matplotlib)

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

# Plot
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Chart Example')
plt.show()
```

## 3-Pie Charts (Matplotlib)

```python
import matplotlib.pyplot as plt

# Data
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

# Plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```

## 4-Scatter Plots (Matplotlib)


```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

# Plot
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot Example')
plt.show()
```
